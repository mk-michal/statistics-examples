### Poisson distribution and process

Poission process examples:

    − the number of car accidentsat a site or in an area;
    − the location of users in a wireless network;
    − the requests for individual documents on a web server;
    − the outbreak of wars;

```python
def poisson_distribution(l, k):
    '''calculates poisson distribution'''
    probability = (l ** k * np.exp(-l))/math.factorial(k)
    return probability

def plot_poisson_distribution(l: int, x_max= 100):
    '''Plots poisson distribution'''
    plt.ylabel('Probability of car passing')
    plt.xlabel('Number of cars')
    plt.title('Probability Distribution Curve')
    arr = []
    for num in range(0,x_max):
        arr.append(poisson_distribution(l, num))

    plt.grid(True)
    plt.plot(arr, linewidth=2.0)
    plt.show()
    
    
def plot_poisson_process(lambdas: List[int] = [1,4,8], n_days: int = 50):
    ''' Plots simple example of poisson process for different values of lambda'''
    poisson = [np.random.poisson(lam, size=n_days) for lam in lambdas]
    poisson_cumsum = [list(np.cumsum(p)) for p in poisson]
    X = np.linspace(0, n_days, n_days)

    # Plot the graph
    graphs = [plt.step(X, poisson_cumsum[i], label = f'Lambda = {lambdas[i]}') for i in range(len(lambdas))]
    plt.legend(handles = [plot[0] for plot in graphs], loc=2)
    plt.title("Poisson Process")
    plt.ylim(0), plt.xlim(0), plt.grid(True)
    plt.xlabel('Number of days'), plt.ylabel('Number of events')
    plt.show()
```

```python
interact(plot_poisson_distribution, l = (0, 50, 1))
```

```python
plot_poisson_process()
```

### Exponential distribution:
It describes the probability of an event occurring between two timeframes. Continuous random variable 

    - The time until a radioactive particle decays, or the time between clicks of a geiger counter
    - The time it takes before your next telephone call
    - The time until default (on payment to company debt holders) in reduced form credit risk modeling

```python
def exponential_probability(lamb: float, time: float):
    ''' Calculates probability of exponential random variable'''
    return lamb * np.exp(-lamb * time)


def plot_exponentail_distribution(lamb: List[float], t_max: int = 10):
    ''' Plots exponential distribution'''
    x = np.linspace(0, t_max, 100)
    y = [[exponential_probability(l, time) for time in x] for l in lamb]
    
    plt.ylabel('Probability of an event'), plt.xlabel('Time'), plt.title('Exponential probability')
    graphs = [plt.plot(x, y_values, label = f' Lambda = {l}') for y_values, l in zip(y, lamb)]
    plt.legend(handles = [plot[0] for plot in graphs], loc=2)

    plt.show()
    
```

```python
plot_exponentail_distribution([2,1, 0.5])
```
